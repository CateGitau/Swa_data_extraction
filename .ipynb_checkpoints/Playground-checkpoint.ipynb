```python
import requests as rq
import bs4 as bs
```

```python
resp = rq.get("https://www.bbc.com/swahili")
```

```python
page_soup = bs.BeautifulSoup(resp.text, 'html.parser')
```

```python
tmp_links = page_soup.select("a")
```

```python
refs = []
for l in tmp_links:
    name = l["href"]
    if name.startswith("/swahili/habari-"):
        refs.append(l["href"])
```

```python
refs
```

```python
results = []
for resource in refs:
    link_url = "https://www.bbc.com/" + resource
    resp = rq.get(link_url)
    page_soup = bs.BeautifulSoup(resp.text, 'html.parser')
    
    title = page_soup.select_one("h1.story-body__h1")
    title = title.text if title else ''
    body = page_soup.select("div.story-body__inner p")
    
    final_text = " "
    for p in body:
        p = p.text if p else ''
        final_text += p
    results.append({"title" : title, "body" : final_text})
    print("Done with : " + resource)
```

```python
results
```

```python
import json
with open('data_trial.json', 'w') as fout:
    json.dump(results, fout)
```

```python
with open("data_trial.json", "r") as read_file:
    data = json.load(read_file)
```
