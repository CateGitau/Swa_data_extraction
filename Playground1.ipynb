# Habari

```python
refs = []
for l in news_links:
    name = l["href"]
    if name.startswith("/swahili/habari-"):
        refs.append(l["href"])
```

```python
refs
```

```python
results = []
for resource in refs:
    link_url = "https://www.bbc.com/" + resource
    resp = rq.get(link_url)
    page_soup = bs.BeautifulSoup(resp.text, 'html.parser')
    
    title = page_soup.select_one("h1.story-body__h1")
    title = title.text if title else ''
    body = page_soup.select("div.story-body__inner p")
    tags = page_soup.select("li.tags-list__tags a")
    
    final_text = " "
    for p in body:
        p = p.text if p else ''
        final_text += p
        
    final_tag = []
    for q in tags:
        q = q.text if q else ''
        final_tag.append(q)
    final_tag = ','.join(final_tag)
    results.append({"title" : title, "body" : final_text, "tags": final_tag})
    print("Done with : " + resource)
```
